## Handeling Outlier by Using IQR Inter Quantile Range


```python
# Identify the first quartile (Q1), third quartile (Q3), and interquartile range (IQR) for the specified columns

Q1 = clean3[column_name].quantile(0.25) # Quantile is a function 
Q3 = clean3[column_name].quantile(0.75)
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers for each column
clean3[column_name] = np.where(clean3[column_name] < lower_bound, lower_bound, clean3[column_name])
clean3[column_name] = np.where(clean3[column_name] > upper_bound, upper_bound, clean3[column_name])
```

```python
# Plot box plots for cleaned data
plt.figure(figsize=(15, 10))
clean3[column_name].plot(kind='box', subplots=True, layout=(1, 9), figsize=(15, 5), sharey=False)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
from sklearn.linear_model import LinearRegression

# split the data set into train and test
x_train,x_test,y_train,y_test = train_test_split (x,y, test_size=0.25,random_state=42)
```

Feature Scalling 

```python
standerd_scaler = StandardScaler()
```

```python
X_train_std = standerd_scaler.fit_transform(x_train)
X_std_test = standerd_scaler.transform(x_test)
```

Linear Regrssion 

```python
# Model selection ---Linear regression 
linear_model = LinearRegression()
linear_model.fit(X_train_std,y_train)
```

```python
# predicted value
L1_pred = linear_model.predict(X_std_test)
```

```python
L1_pred
```

```python
print('intercept',linear_model.intercept_)

print('coeff',linear_model.coef_)
```

```python
## Cross validation 
from sklearn.model_selection import cross_val_score
cross_val = cross_val_score(linear_model,x,y,cv=5)

print(cross_val)
```
